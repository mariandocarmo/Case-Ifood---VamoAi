```python
columns = ['Type', 'App Id', 'App Website',
       'Metascore', 'Metascore Link', 'Release Date',
       'Developers', 'Publishers', 'Windows', 'Mac', 'Linux',
       'Install Size (MB)', 'Controller Support', 
       'Has Steam Cloud', 'Has Achievements', 'Has Trading Cards',
       'Has Steam Workshop', 'Has Stats', 'Has HDR']
```

### Vamos tirar algumas colunas.

```python
df.drop(columns=columns, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Has Multiplayer'].unique()
```

```python
df['Has Multiplayer'] = df['Has Multiplayer'].replace(np.nan, 'False')
```

```python
df.head()
```

### Alguns jogos na biblioteca que nunca joguei

```python
df['Time Played (minutes)'].value_counts()
```

```python
df = df[df['Time Played (minutes)'] > 1]
```

```python
df.sample(5)
```

```python
df.rename(columns={'Title': 'nome', 'Time Played (minutes)': 'tempo_jogado_min', 
                   'Time To Beat 100% (minutes)': 'tempo_para_finalizar_min', 
                  'Time To Beat Main Game (minutes)': 'tempo_para_finalizar_jogo_principal_min', 
                 'Regular Price (USD)': 'preço_em_real', 'Has Multiplayer': 'eh_multiplayer',
                  'Genres': 'generos'}, inplace=True)
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns='index', inplace=True)
```

```python
df.head()
```

```python
df.tempo_jogado_min.sort_values(ascending=False).head(10)
```

```python
df.sort_values(by='tempo_jogado_min', ascending=False).head()
```

```python
df['tempo_jogado_hora'] = df['tempo_jogado_min'] / 60
```

```python
df.sort_values(by='tempo_jogado_hora', ascending=False).head()
```
